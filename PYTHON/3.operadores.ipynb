## Operadores aritméticos


```python
a = 10
b = 5

print(a + b)
print(a - b)
print(a * b)
print(a / b)
print(a // b)
print(a % b)
print(a**b)

# La precedencia funciona exactamente igual que en las matemáticas cotidianas
print(2 + 3 * 4)
print((2 + 3) * 4)

# Operadores de reasignación
c = 5

# c = c + 3
c += 3
print(c)
```

## Operadores de comparacion


```python
x = 10
y = 3

print(x == y)
print(x != y)
print(x > y)
print(x >= y)
print(x < y)
print(x <= y)

resultado = x == y
print(type(resultado))
```

```python
print("Ana" == "ana")
print("Ana" < "Andres")

print("Berto".lower() == "BERTO".lower())
```

## Operadores logicos y pertenencia


```python
edad = 25
tiene_carnet = False

print(edad >= 18 and tiene_carnet)
print(edad >= 18 or tiene_carnet)

print(not tiene_carnet)
```

```python
print(True and True)
print(True and False)
print(False and False)

print(True or False)
print(False or False)
```

```python
nota = 7
asistencia = 85

print((nota >= 5) and (asistencia >= 80))
```

```python
frase = "Estamos aprendiendo Python"

print("Python" in frase)
print("Java" not in frase)
print("python".capitalize() in frase)
```
