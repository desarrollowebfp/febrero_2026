## Indexación


```python
lenguaje = "Python"

print(lenguaje[0])
print(lenguaje[3])
print(lenguaje[-1])
print(lenguaje[-2])
```

## Slicing


```python
string = "Eso es una frase"

# [start:stop:step]
print(string[11:])
print(string[0:3])
print(string[:3])
print(string[:])
print(string[4:])
print(string[0:3:1])
print(string[::2])
print(string[::-1])
```

## Métodos de verificación


```python
print("12345".isdigit())
print("abced".isalpha())
print("abc123".isalnum())
print("    ".isspace())
```
