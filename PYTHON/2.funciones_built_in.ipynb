## Funciones built-in generales


```python
# Mostrar algo por terminal o pantalla
print("Hola")
# Mostrar el tipo de dato
edad = 30
print(type(edad))
# Mostrar la longitud de caracteres o elementos
print(len("Python"))

# Permitir entrada de usuario
# nombre = input("¿Cómo te llamas?")
# print(nombre)

print(type(int("30")))
print(type(float("30.5")))
print(type(str(40)))

print(bool(0))
print(bool(-1))

print(round(4.7))
```

## Funciones built-in para strings


```python
frase = "     Python es el MEJOR lenguaje de programación.          "
print(frase)

# Eliminar los espacios sobrantes al principio y al final del string
print(frase.strip())
frase = frase.strip()
# Ponemos todo en mayuscula
print(frase.upper())
# Ponemos todo en minúscula
print(frase.lower())
# Ponemos la primera letra en mayúscula
print("hola, esto es un string".capitalize())
# Poner la primera letra de cada palabra en mayúscula
print(frase.title())
# Se pueden combinar varios metodos seguidos
print(frase.lower().capitalize())
# Sustituir un sub-string por otro
print(frase.replace("Python", "JavaScript"))
# Hacer un recuento de apariciones de un substring
print(frase.count("a"))
# Comprobar con un booleano si un string termina en un substring
print(frase.startswith("Py"))
print(frase.endswith("ón."))


# Ejemplo de normalización de datos
nombre = input("¿Cuál es tu nombre")
nombre = nombre.strip().title()

print("El nombre del usuario es: " + nombre)
```
